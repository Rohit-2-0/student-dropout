Zoomed-In Correlation Heatmap

Code Summary:

    Calculates the absolute correlation between all features and the encoded target variable (Target_encoded).

    Selects features that have a correlation greater than 0.3 with the target.

    Builds a filtered correlation matrix using these top features.

    Plots a heatmap to visually analyze the interrelationships.

The heatmap highlights the strongest predictors of the target (e.g., dropout likelihood).

Some key features strongly correlated with the target may include:

    Grades in 1st and 2nd semesters (e.g., Curricular units 1st sem (grade), Curricular units 2nd sem (grade)).

    Academic performance metrics like approved/evaluated subjects.

    Tuition fees payment status or debtor status.

These features can be considered for feature selection in predictive modeling.

```python
# Step 1: One-hot encode the target
df_encoded[['Class_0', 'Class_1', 'Class_2']] = pd.get_dummies(df_encoded['Target_encoded'])

# Step 2: Drop encoded target to avoid confusion
df_corr = df_encoded.drop(columns=['Target_encoded'])

# Step 3: Get correlation with each class column
corr_matrix = df_corr.corr()


# Step 4: Correlation with each class, dropping the target columns
corr_class_0 = corr_matrix['Class_0'].drop(['Class_0', 'Class_1', 'Class_2'])
corr_class_1 = corr_matrix['Class_1'].drop(['Class_0', 'Class_1', 'Class_2'])
corr_class_2 = corr_matrix['Class_2'].drop(['Class_0', 'Class_1', 'Class_2'])

# Step 5: Get top correlated features (> 0.3 abs value)
top_class_0 = corr_class_0[abs(corr_class_0) > 0.3].sort_values(ascending=False)
top_class_1 = corr_class_1[abs(corr_class_1) > 0.3].sort_values(ascending=False)
top_class_2 = corr_class_2[abs(corr_class_2) > 0.3].sort_values(ascending=False)

# Step 6: Plot function
def plot_heatmap(features, class_name):
    if features.empty:
        print(f"No features with correlation > 0.3 for class {class_name}")
        return
    corr_matrix = df_encoded[features.index].corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True, linewidths=0.5)
    plt.title(f"Correlation Heatmap - Top Features for {class_name}", fontsize=16)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

# Step 7: Plot
plot_heatmap(top_class_0, 'Class 0 (Enrolled)')
plot_heatmap(top_class_1, 'Class 1 (Dropout)')
plot_heatmap(top_class_2, 'Class 2 (Graduated)')
# Step 8: Save the processed DataFrame
```

```python
# --- Model Training and Evaluation ---
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score,roc_curve, auc

from sklearn.preprocessing import label_binarize
import numpy as np

X_selected = X
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42, stratify=y)

models = {
    "RandomForest": RandomForestClassifier(),
    "DecisionTree": DecisionTreeClassifier(),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric="mlogloss")
}

classes = np.unique(y)
y_test_bin = label_binarize(y_test, classes=classes)

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)

    print(f"🔍 Model: {name}")
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

    if len(classes) > 2:
        print("ROC AUC Score:", roc_auc_score(y_test_bin, y_proba, multi_class='ovr'))
    else:
        print("ROC AUC Score:", roc_auc_score(y_test, y_proba[:, 1]))

    print("-" * 50)
```

```python

# Model Names
models = ['Random Forest', 'Decision Tree', 'XGBoost']

# Performance Metrics (replace with your exact results if needed)
accuracy = [0.83, 0.75, 0.83]
precision = [0.80, 0.74, 0.81]  # for dropout (class 1)
recall = [0.87, 0.78, 0.86]
f1_score = [0.83, 0.76, 0.83]
roc_auc = [0.9036, 0.7548, 0.9041]

# Organize data
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC AUC']
metric_values = [accuracy, precision, recall, f1_score, roc_auc]

# Plotting
bar_width = 0.15
x = np.arange(len(models))
plt.figure(figsize=(12, 6))

for i, metric in enumerate(metrics):
    plt.bar(x + i * bar_width, metric_values[i], width=bar_width, label=metric)

plt.xticks(x + bar_width * 2, models)
plt.ylim(0.6, 1.0)
plt.ylabel("Score")
plt.title("✅ Model Performance Comparison")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

for name, model in models.items():
    model.fit(X_train, y_train)
    y_proba = model.predict_proba(X_test)

    if len(classes) == 2:
        fpr, tpr, _ = roc_curve(y_test, y_proba[:, 1])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2,
                 label=f"{name} (AUC = {roc_auc:.2f})")
    else:
        y_test_bin = label_binarize(y_test, classes=classes)
        for i in range(len(classes)):
            fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_proba[:, i])
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, lw=2,
                     label=f"{name} - Class {classes[i]} (AUC = {roc_auc:.2f})")

plt.plot([0, 1], [0, 1], linestyle='--', color='gray', lw=2)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves")
plt.legend(loc="lower right", fontsize='small')
plt.grid(True)
plt.tight_layout()
plt.show()
```
